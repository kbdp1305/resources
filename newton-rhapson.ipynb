# NOMOR 1

```python
pi=math.pi
def fungsi(h) :
    return pi*h**3-9*pi*h**2+90
    
```

```python
def turunan(h) :
    return 3*pi*h**2-18*pi*h
```

```python

def newton(x1) :
    x0=0
    x_1=[]
    x_0=[]
    
    while(abs(x1-x0)>0.00000001) :
       
        x0=x1
        x_0.append(x0)
        x1=x0-(fungsi(x0)/turunan(x0))
        x_1.append(x1)
    return x1,x_0,x_1
    
```

Kita inisiasi dengan x awal = 3 sebab jika x awal =0 , maka akan terjadi eror

```python
x,x_0,x_1=newton(3)
```

```python
df=pd.DataFrame()
df['h']=x_1
df['x0']=x_0
```

```python
def function2(h) :
    return 15*h**2+5*h**3-16
```

```python
def turunan2(h) :
    return 15*h**2+30*h
```

```python

def newton2(x1) :
    x0=0
    x_1=[]
    x_0=[]
    
    while(abs(x1-x0)>0.00000001) :
       
        x0=x1
        x_0.append(x0)
        x1=x0-(function2(x0)/turunan2(x0))
        x_1.append(x1)
    return x1,x_0,x_1
```
